```python
"""
- Supported languages:
  am: Amharic
  ar: Arabic
  ff: Fulah/Fulfulde
  ha: Hausa
  ig: Igbo
  lg: Ganda
  ln: Lingala
  mg: Malagasy **
  ns: Northern Sotho **
  sw: Swahili **
  so: Somali **
  ss: Swati **
  tn: Tswana **
  xh: Xhosa **
  yo: Yoruba **
  zu: Zulu **

  ** - Languages supported by M2M100 but not considered in the POV(27 languages)
"""
# TARGET_LANG = ['am', 'ar', 'ff', 'ha', 'ig', 'lg', 'ln']
dataset_path = '/content/drive/MyDrive/flickr8k_best_captions.csv'
```

```python
from transformers import AutoTokenizer, M2M100ForConditionalGeneration
import pandas as pd
```

```python
# Load model and Tokenizer
m2m_model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
m2m_tokenizer = AutoTokenizer.from_pretrained("facebook/m2m100_418M")
```

```python
# Read file
df = pd.read_csv(dataset_path, header=0)
df.head()
```

```python
target_lang = "yo"

translated_captions = []

for index, row in df.iterrows():
  caption = row['best_caption']

  # Tokenize/Encode orginal caption (in English)
  model_inputs = m2m_tokenizer(caption, return_tensors="pt")

  # Pass our tokenized sentence into the model
  gen_tokens = m2m_model.generate(**model_inputs, forced_bos_token_id=m2m_tokenizer.get_lang_id(target_lang))

  # Decode output of the model
  translated_text = m2m_tokenizer.batch_decode(gen_tokens, skip_special_tokens=True)[0]

  translated_captions.append({
        "image": row['image'],
        "en": caption,
        target_lang: translated_text
    })

translated_captions
```
